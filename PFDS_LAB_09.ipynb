```python
# Lab 9: Visualize Relationships Between Variables Using Seaborn Pair Plots
# Scenario 4: Housing Data Analysis

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("housing_data.csv")

# Display first few rows
print("Sample Data:")
print(df.head())

# Customize Seaborn theme and color palette
sns.set(style="whitegrid", palette="coolwarm", font_scale=1.1)

# Select numerical features for visualization
selected_features = ['Price', 'Area_sqft', 'Rooms', 'Age', 'Distance_to_City_km']

# Create a pair plot
pair_plot = sns.pairplot(df[selected_features], diag_kind='kde', corner=True, height=2.5)

# Add main title
pair_plot.fig.suptitle("Pair Plot of Housing Features", y=1.02, fontsize=14, weight='bold')

# Show plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Filter out rows with missing data
titanic_clean = titanic[['age', 'fare', 'pclass', 'survived']].dropna()

# Save to CSV
titanic_clean.to_csv("titanic_clean.csv", index=False)
print(" Cleaned Titanic dataset saved as titanic_clean.csv")

# Create pair plot
sns.pairplot(titanic_clean, hue="survived", diag_kind="kde", palette="Set2")
plt.suptitle("Titanic Passenger Patterns: Age, Fare, and Class", y=1.02)
plt.show()
```
