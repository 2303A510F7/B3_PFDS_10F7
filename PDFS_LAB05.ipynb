Create/import two datasets (Products and Sales)

```python
import numpy as np
import pandas as pd
p = pd.read_csv('/content/products.csv')
p.head()
```

```python
s=pd.read_csv('/content/sales.csv')
s.head()
```

Merge them using product_id

```python
merge=pd.merge(p,s,on='product_id')
merge.head()
```

Perform inner and outer joins

```python
inner_join=pd.merge(p,s,on='product_id',how='inner')
inner_join.head()
```

```python
outer_join=pd.merge(p,s,on='product_id',how='outer')
outer_join.head()
```

Compute statistics like total revenue, average price, and
product-wise sales.

```python
total_revenue=merge['price'].sum()
print("Total revenues:",total_revenue)
avergae_price=merge['price'].mean()
print("Average price:",avergae_price)
product_wise_sales=merge.groupby('product_id')['price'].sum()
print("Product wise sales:",product_wise_sales)
```

Group by product category and find best-selling category

```python
best_selling_category=merge.groupby('category')['price'].sum().idxmax()
print("Best selling category:",best_selling_category)
```

Create two datasets (Teachers and Courses)

```python
t=pd.read_csv('/content/teachers.csv')
t.head()
```

```python
c=pd.read_csv('/content/courses.csv')
c.head()
```

Merge on teacher_id

```python
merge=pd.merge(t,c,on='teacher_id')
merge.head()
```

Perform inner and left joins

```python
ij=pd.merge(t,c,on='teacher_id',how='inner')
ij.head()
```

```python
oj=pd.merge(t,c,on='teacher_id',how='left')
oj.head()
```

Compute average courses per teacher

```python
average=merge.groupby('teacher_id')['course_id'].count().mean()
print("Average courses per teacher:",average)
```

Group by department and calculate total workload

```python
workload=merge.groupby('department')['course_id'].count().sum()
print("Total workload:",workload)
```
